# Inheritance in python OOP

```python
class student: #base class
    def dept(self):
        print(f"dept = ICE")

class yahiya(student): #child class
    def display(self):
        print(f"Roll : 12")
        print(f"Bathch : 15th")
class kamal(student):   #child class
    pass

y =  yahiya()
k = kamal()

y.dept()
y.display()
print()

k.dept()

```

# overriding

```python
"""
Method Overriding meansa child class defines a method
with the same name as the parent class.
The child's version replaces the parent's version.

"""
```

```python
class human:
    def intro(self):
        print("i'm human!")
class men(human):
    def intro(self):
        print("i'm men")
class women(human):
    pass

m = men()
w = women()

m.intro()
w.intro()
```

# super() function

```python
# age parent method then child method

class student:
    def info(self):
        print("Dept : ICE")
class yahiya(student):
    def info(self):
        super().info()
        print("i'm yahiya")

y = yahiya()
y.info()
```

```python
# only parent method

class student:
    def info(self):
        print("Dept : ICE")
class yahiya(student):
    def info(self):
        super().info()

y = yahiya()
y.info()
```

# super() with constructor

```python
# constructor in parent and child both will work

class student:
    def __init__(self):
        print("Dept : ICE")
        
class yahiya(student):
    def __init__(self):
        print("i'm yahiya")
        super().__init__()

y = yahiya()
```

# many types of inheritance

**Multiple Inheritance**

```python
class a:        # parent 1
    def a(self): print("a")

class b:        # parent 2
    def b(self): print("b")

class c(a,b):
    pass

child = c()
child.a()
child.b()
```

**Multilevel** _______________ **Grandparent → Parent → Child**

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

**Hierarchical**

```python
"""
        Parent
       /      \
   Child1   Child2

"""
```

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

**Hybrid**

```python
"""
         A
       /   \
      B     C
       \   /
         D

"""
```

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass
```
