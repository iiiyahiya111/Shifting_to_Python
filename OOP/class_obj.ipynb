## Class and Object

```python
class student:
    pass

stu1 = student()   # stu1 is an object of the class "student"
stu2 = student()   # stu2 is another object of the class "student"


# assign value to these objects
stu1.name = "Alice"
stu1.age = 20

stu2.name = "Bob"
stu2.age = 22

print(stu1.name, stu1.age)
print(stu2.name, stu2.age)

```

```python
print(type(stu1))
```
