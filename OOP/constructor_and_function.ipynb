# Function in class

```python
class student:
    def __init__(self, n, r):
        self.name=n     #can be same like: self.name = name
        self.roll=r
    def display(self):
        print(f"{self.name} : {self.roll}")

stu1 =  student('yahiya', 1234)
stu1.display()
```

```python
class student:
    def __init__(self, n, r):
        self.name=n     #can be same like: self.name = name
        self.roll=r
    def display(self, subj):
        print(f"{self.name} : {self.roll}")
        print(f"Subject name : {subj}")     

stu1 =  student('yahiya', 1234)
stu1.display('Programing')
```

```python
class student:
    highest_cg = 2.9
    def __init__(self, n, r):
        self.name=n     #can be same like: self.name = name
        self.roll=r
    def display(self):
        print(f"{self.name} : {self.roll}")
    def update_cg(self,x):
        # reference the class attribute via self or the class name
        print(f"Current CG is : {self.highest_cg + x}")

stu1 =  student('yahiya', 1234)
stu1.display()
stu1.update_cg(1)
```
