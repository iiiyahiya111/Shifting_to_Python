```python
def student(name,dept):
    print(f"Hi! i'm {name}, and i'm a {dept} guy...")

student(dept="CSE",name="Barakat")
```

**Default Argument**

```python
def student(name="Yahiya",dept="ICE"):
    print(f"Hi! i'm {name}, and i'm a {dept} guy...")

student()
student("CSE","Barakat")
student(dept="CSE",name="Barakat")
#student(name="Barakat","cse") #Not Allowed
student("Barakat",dept="EEE")

```

**With return**

```python
def sum(a,b):
    return a+b
x=sum(3,4)
print(x)
```

##**Variable-Length Arguments / Arbitrary Arguments**

```python
def sum(*a):
    s=0
    for i in a:
        s=s+i
    print(f"Sum is: {s}")

sum(1,2,3,4)
```
