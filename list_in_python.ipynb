built in Function for list

```python
num2=[4,1,2,5,3]
num2.reverse()
print("revers: ")
print(num2)


num2.sort()
print("sort:")
print(num2)


num2.append(99) #end of the list
print("append(end of list)")
print(num2)


num2.insert(0,100) #specific position
print("insert(specific position)")
print(num2)


num2.extend([66,77,88])
print("extend(append multiple end of the list)")
print(num2)
```

Change/replace any Value

```python
num2=[4,1,2,50,3]

num2[1]=100
print(num2)

num2[1:3]=[22,23]
print(num2)

num2.remove(50) #remove the value 50
print(num2)


print(num2.pop())
print(num2)
```

Revers Print

```python
num2=[4,1,2,50,3]

print(num2[0::-1])
```
