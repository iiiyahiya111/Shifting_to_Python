# Dictionary in python

Dictionary is ordered (py - 3.7), keys are immutable but value are not, value can be changed.
Duplicate keys are not allowed.

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    1: 'jarif'
}

print(student)
```

**Access the value**

```python
print(student[1])
```

**Another Dict**

```python
roll ={
    'yahiya':1,
    'abid' : 2,
    'asif' : 3,
}

print(roll['yahiya'], roll['asif'])
```

**Another method - using dict() function.**

```python
name_=dict(
    {
    'yahiya':1,
    'abid' : 2,
    'asif' : 3,
}
)
print(name_)
```

**Its also can be** *::*

```python
stu = dict([
    ('yahiya',11),
    ('abid' , 22),
    ("akif" , 33)
])
print(stu)
```

**Changing the value**

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    1: 'jarif'
}
print(student)
student[2]=123
print(student)

student[2]=student[2]+1
print(student)
```

**add value after declaration**

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    4: 'jarif'
}
student['masum']=123

print(student)
```

**Nested Dict**

```python
phone_no={
    'yahiya' : 12,
    'abid' : 234,
    'afif' : 980,
    'jarif' : { 'j1' : 222, 'j2' : 333 }, 
}
print(phone_no)
```

**get() method**

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    'Yahiya' : 1234
}

print(student.get(3))
print(student['Yahiya'])
#print(student['yahiya']) #get error
print(student.get('yahiya'))
```

**Delete function, del and pop**

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    'Yahiya' : 1234
}
del student['Yahiya']
print(student)

print(student.pop(3)) #pop() function remove and return the specific value
```

```python
student = {
    1 : 'yahiya',
    2: 'abid',
    3: 'arif',
    'Yahiya' : 1234
}
print(student.popitem()) #delete last items and return the value pair
```

```python
print(student.keys()) #only shows the keys
print(student.values()) #only value
print(student.items())
```

# For loop in dict

```python
for i in student:
    print(i) #print keys only
```

```python
for i in student:
    print(student[i]) #only value
```

```python
for i in student.items():
    print(i)
```
