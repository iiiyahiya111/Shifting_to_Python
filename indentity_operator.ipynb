
Identity operator: is, is not

```python
a=5
b=5     #use the same data as a instead of create another object name b
print(a is b) #same object, same id just different in the tag/name of object
#output will be True

print(a is not b)
#output will be false

print(id(a))
print(id(b))
```

```python
a=5
b='5'
print(a is b)
print(a is not b)

print(id(a))
print(id(b))
```

```python
# Objects like ints and strings are immutable: operations create new objects
print("before a:", a, "id:", id(a))
a = a + 1
print("after a:", a, "id:", id(a))

print("before b:", b, "id:", id(b))
b = b + "0"
print("after b:", b, "id:", id(b))
```
